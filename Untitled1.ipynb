```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("/content/tmdb_5000_movies.csv",skiprows = [4298])
credits = pd.read_csv("/content/tmdb_5000_credits.csv",skiprows = [433])
```

```python
credits.head()
```

```python
movies.head()
```

```python
# we will join the two dataset based on the title column
movies = movies.merge(credits,on = "title")
```

```python
movies.head(1)
```

```python
# we will keep the columns which we need in the dataset

movies = movies[["movie_id","title","overview","genres","keywords","cast"]]
```

```python
movies.head(1)
```

```python
# we will create three columns and create a new columns and merge the column from the overview to the end

# check if the datasert has null values
movies.isnull().sum()
```

```python
# check if there are duplicate data
movies.duplicated().sum()
```

```python
# convert the string list to a proper list
import ast
ast.literal_eval
```

```python
# we will make the genre column in the proper list from string to list
def covert(obj):
  L = []
  for i in ast.literal_eval(obj):# helps to convert the string into the list
    L.append(i["name"])
  return L
```

```python
movies["genres"] = movies["genres"].apply(covert)
```

```python
movies.head()
```

```python
# we will do the same thing on the keywords column
movies["keywords"] = movies["keywords"].apply(covert)
```

```python
movies.head()
```

```python
# for the cast we will add the top three actors
def covert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):# helps to convert the string into the list
    if counter != 3:
      L.append(i["name"])
      counter+= 1
    else:
      break
  return L
```

```python
movies["cast"] = movies["cast"].apply(covert3)
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x : [i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x : [i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# we need to convert the overview into a list to concatenate with the list
movies["overview"] = movies["overview"].apply(lambda x :x.split())
```

```python
# we will create a tag column
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
# convert the list into a string
new_df["tags"] = new_df["tags"].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
# vectoriation on the tags
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000 , stop_words ="english")

```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
#we need to do steaming import
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

    # conerting string to the list
  return " ".join(y)


```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
# we need to find the cosine distance between two vectors it gives the angle between the vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# recommending the first 5 movies
def recommend(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  # need to find the distances
  distance = similarity[movie_index]
  movies_list = sorted(list(enumerate(distance)),reverse = True, key = lambda x : x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)



```

```python
recommend("Batman Begins")
```
